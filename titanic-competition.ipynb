# Model Building:

```python
x_train=train[['Age']]
y_train=train[["Survived"]]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
sum(test['Age'].isnull())
```

```python
test=test.dropna()
sum(test['Age'].isnull())
```

```python
x_test=test[['Age']]
y_test=dtc.predict(x_test)
```
